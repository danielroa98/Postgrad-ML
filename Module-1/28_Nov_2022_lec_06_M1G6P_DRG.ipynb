#### Para emular un `do while`

Utilizar un `break` cuando se desea romper un ciclo antes de la siguiente iteración.

Ejemplo:

Guardar una lista de libros que nos gusten, require input del usuario.

```python
books = []
print("Type in a book's title or type in 'fin' to finish the execution")


while True:
    title = input('Write the title or "fin": ')
    if title != 'fin':
        books.append(title)
    else:
        print('Stopping...')
        break

print(f'The list contains {books}')
```

## Ejercicios

1) Dada una lista de enteros, obtener el máximo de estos.

```python
numbers = [-9,-23,7,2,6,8,-34]
```

```python
def bigger_num(val_list):
    maxn = val_list[0]
    for i in range(len(numbers)):
        if maxn < val_list[i]:
            maxn = val_list[i]
    return maxn
```

```python
print(bigger_num(numbers))
```

```python
# Esta versión es en caso que una lista contenga solo ints.
def bigger_numV2(val_list):
    maxn = val_list[0]
    for i in numbers:
        if maxn < i:
            maxn = i
    return maxn
```

```python
print(bigger_numV2(val_list=numbers))
```

2) Dada una lista, imprimir la suma de valores de la lista

```python
def sum_of_values(val_list):
    total = 0
    for i in val_list:
        total = i + total
    return total
```

```python
test_list = [1,2,3,4,5]
print(sum_of_values(test_list))
```

3) Pedir al usuario un numero y calcular la sume de los números entre 1 y el número que ingrese.

```python
total_sum = 0

n = int(input('Enter a positive number: '))

for i in range(1, n+1):
    total_sum += i

print(total_sum)
```

4) Calcular el factorial de un número (entero no repetitivo)

```python
factorial = 1
n = int(input("Ingrese un número entero: "))

if n<0:
    print("Error, no se pueden ingresar números negativos")
elif n==0:
    print("0 factorial equals to 1")
else:
    for i in range(1, n+1):
        factorial = factorial*i

print(f'The factorial of {n} is {factorial}')
```

### Agrupando elementos con `zip`

```python
tick = ['APPL','AMZN','META','GOOG']
companies = ['Apple', 'Amazon','Meta','Alphabet']

# Diccionario para guardar los valores
comp_ticker = {}

for ticker, company in zip(tick, companies):
    comp_ticker[company] = ticker

print(comp_ticker)
```

### Lambda functions

Ejemplo sencillo

- Los parámetros _x, y_ se declaran después de la palabra reservada `lambda`.
- Después de los dos puntos (:), se indica que operación realizará la función.
- Por último, le pasamos los valores de _`x`_ y _`y`_.

```python
(lambda x,y: x**y)(2,3)
```

```python
# Versión expandida del lambda function anterior
def potencia(x,y):
    return x**y
```

```python
potencia(2,3)
```

#### Las funciones lambda permiten pasar funciones como params SIN haberlos definidos previamente
Composición de funciones `fog(x) = f(g(x))`

Se define la función f con dos argumentos: _g (la función)_ y _x_.

```python
def f(g, x):
    return 2*g(x)

# En la función f se está mandando una función lambda como g
## i.e.: x**(1/2) y x=3
## por lo tanto, si x=3, g(x) = sqrt(3)*2 = 3.46...
f(lambda y:y**(1/2), 3)
```

## Función `filter()`

Nos permite filtrar elementos de una secuencia (listas, conjuntos, tuples, etc...), a través de una función que cumpla con cierto criterio.

Sintaxis:

`filter(f(), sequence)` -> devuelve los elementos de la secuencia que son `True` según las condiciones impuestas por la función `f()`.
