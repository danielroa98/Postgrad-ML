### ```dropna()```
Para eliminar renglones que tienen __ALGÚN__ NaN (o valor nulo) en cualquier variable.

- El valor default de how es 'any' por lo tanto elimina los renglones que tengan 1 o mas valores nulos en sus columnas.

````python
nba.dropna()
````

Si queremos quitar los rengoles que tienen NaN en __todas__ las columnas

````python
nba.dropna(how='all')
````

Se quitó solo un renglón, debido a que solo había un renglón que tenía mas valores como NaN

````python
nba.tail(3)
````

Si se quieren quitar los renglones de forma permanente, solo se tiene que añadir el atributo ```inplace=True```.

````python
nba.dropna(how='all', inplace=True)
````

````python
nba.tail(3)
````

````python
nba.info()
````

Eliminar columnas con al menos un valor nulo

````python
# El 1 es para las columnas
nba.dropna(axis=1)
# Quitó las columans College y Salary debido a que estas tienen al menos un valor nulo.
````

````python
# Equivalente a lo anterior
nba.dropna(axis='columns')
````

Se quiere el número de registros no nulos en la variable _College_

````python
# Mi manera
nba['College'].notnull().value_counts()

# Manera vista en clase
# nba['College'].notnull().sum()
````

Se quiere eliminar cualquier columna que tenga todos los valores nulos

````python
nba.dropna(how='all', axis=1)
````

````python
nba['Nulls'] = np.nan
nba.head()
````

Se elimina la columna donde todos los valores que contenía eran nulos.

````python
nba.dropna(axis=1, how='all', inplace=True)
````

Se quieren afectar columnas específicas
- Elimina filas que tengan al menos un nulo en las variables especificadas.

````python
nba.dropna(subset=['College', 'Salary'])
````
