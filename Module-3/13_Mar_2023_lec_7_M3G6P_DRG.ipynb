# Ejemplo - obteniendo los datos de una página de BBVA

```python
url = 'https://www.bbva.mx/personas/productos/inversion/normativa-de-fondos/fondos-renta-variable.html'
```

```python
res = requests.get(url, headers=head)
parser = html.fromstring(res.text)
parser
```

```python
# Nombres
nombres = parser.xpath("//tbody/tr/td[1]//div/text()")
nombres
```

Uso de _list comprehension_ para limpiar la lista que obtuvimos

Pasos usados:
1. Se usó la función `replace()` para poder sustituir el salto de línea `\n` por espacios en blanco.
2. Se usa la función `strip()` para quitar los espacios (" ") que aparecen.
3. Se usa la función filter para quitar por completo los espacios vacíos.
4. Se convierte a `list()` de nuevo para que se puedan obtener los valores como `string`s de nuevo.

```python
fondos = list(filter(None, [x.replace("\n","").strip() for x in nombres]))
fondos
```

```python
# Alternativa para llevar a cabo lo mismo pero programado de otra manera
# fontot= [x for x in list(map(lambda x: x.replace("\n","").strip(),fondos)) if x!='']
```

Extracción de la columna _Prospecto_, que contiene los links.

```python
fondos_url = pd.DataFrame(parser.xpath("///tbody/tr/td[2]//a/@href"))
fondos_url
```
