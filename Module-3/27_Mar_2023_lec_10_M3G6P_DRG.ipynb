Predicción de palabras

```python
import numpy as np
import pandas as pd
from nltk.tokenize import RegexpTokenizer
```

```python
cuento = []

poePath = "/Users/danielroa/Programming/Postgrad-ML-lecturer notes/Module-3/Clases/Clase10/poe.txt"

with open(poePath, "r", encoding="utf8") as archivo:
    for linea in archivo:
        cuento.append(linea)
text = " ".join(cuento).replace("\n", " ")
```

```python
tokenizer = RegexpTokenizer(r'\w+')
clean_text = np.array(tokenizer.tokenize((text.lower())))
```

```python
clean_text
```

```python
long_string = ""
for i in clean_text:
    long_string = long_string+" "+i
```

```python
long_string
```

### $P(w_i)=count(w_i)/N$

```python
freq = []
words = []

for word in set(clean_text):
    freq.append(long_string.count(" "+word+" "))
    words.append(word)

freq = list(freq/np.sum(freq))
freq
```

```python
pd.DataFrame({
    "Palabra":words,
    "Freq": freq
}).sort_values("Freq", ascending=False)
```

#### $P(w_i|w_i-1)=count$

```python
from nltk.util import ngrams
```

```python
def bigram_probs(initial, words, clean_text, n=2):

    denominador = len([x for x in clean_text if x == initial])
    ngram = list(ngrams(clean_text, n))
    probs = []
    for i, x in enumerate(words):
        bigram = (initial, x)
        counter = 0

        for j, k in enumerate(ngram):
            if k == bigram:
                counter += 1
            else:
                continue
        probs.append(counter)
    final = np.array(probs)/denominador
    return final
```
