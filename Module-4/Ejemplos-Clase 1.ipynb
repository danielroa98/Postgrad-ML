## Ejemplo: bienes raíces

Entrenamiento supervisado

```python
from sklearn.linear_model import LinearRegression
```

```python
# Creación de una matriz de diseño X y un vector "y" de precios de casas simulados
X = [[100,2],[150,3],[200,4],[250,5]]
y = [150000,200000,250000,300000]
```

```python
# Creación de un objeto de modelo de regresión lineal 
# model = LinearRegression().fit(X,y)
model = LinearRegression(fit_intercept=True).fit(X,y)
```

```python
# Se imprimen los coefficientes del modelo
print(f"Beta 0 {model.intercept_}")
print(f"Beta 1 y Beta 2 {model.coef_}")
```

```python
# Predecimos el precio de una casa con 100 metros cuadrados y 3 habitaciones
precio_pred = model.predict([[180,3]])
print(f"Precio de una casa con 180m3 y 3 habitaciones {round(precio_pred[0], 3)}")
```

## Ejemplo: agrupamiento de especies de iris

Entrenamiento no supervisado

```python
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
sns.set_theme()
```

```python
# Cargando la base de datos de iris
iris = load_iris()
X = iris.data
```

```python
# Definición del número de grupos a generar
k = 5
```

```python
# Creación del modelo de k-means y ajustamos a los datos
kmeans = KMeans(n_clusters=k, random_state=0, n_init=10).fit(X)
```

```python
# Obtenemos los centroides de los grupos generados
centroids = kmeans.cluster_centers_
```

```python
# Obtenemos la etiqueta del grupo asignado a cada flor iris
# Nombre de las clases
labels = kmeans.labels_
```

```python
# Grafiquemos los datos y los centroides de los grupos
plt.figure(figsize=(10,6))
plt.scatter(X[:,0], X[:,1], c=labels, cmap="viridis")
plt.scatter(centroids[:,0],centroids[:,1], marker='*',s=300,c='r')
plt.title("Agrupamiento de especies de iris con KMeans")
plt.xlabel("Longitud del sépalo")
plt.ylabel("Ancho del sépalo")
plt.show()
```

Los _puntos_ amarillos contienen la mayor cantidad de características entre ellos.
- El punto morado que se encuentra en el amarillo puede que tenga las mismas características pero hay algún atributo que no "encaja" que lo cataloga como morado.
