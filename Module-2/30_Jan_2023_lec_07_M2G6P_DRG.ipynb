## Log-Normal

```python
mu = 0
sigma = 1
x = np.linspace(-0.25, mu+4*sigma, 100)
f_x = stats.lognorm(s=sigma, scale=np.exp(mu)).pdf(x)

plt.plot(x, f_x)
plt.title("Distribución Log-Normal teórica")
plt.xlabel("Res")
plt.ylabel("Probabilidad")
plt.show()
```
