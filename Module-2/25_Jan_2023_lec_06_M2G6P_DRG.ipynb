## Distribuciones discretas

#### Bernoulli

_Objetivo_: Modelar un experimento aleatorio con respuesta binaria Ãºnica, como una pregunta de _si o no_.

```python
# 
N = 1000
# tiras una moneda
p = 0.5

# rvs = random variables
## random_state == random_seed, en otras palabras, representa segundos aleatorios.
muestra = stats.bernoulli.rvs(p, size=N, random_state=None)

mu, sigma2 = stats.bernoulli.stats(p, moments="mv")
print(f"Valores estimados")
print(f"E[X]:{mu}, V[X]:{sigma2}, sd[X]:{np.sqrt(sigma2)}")

plt.hist(muestra)
plt.title("DistribuciÃ³n de Bernoulli")
plt.xlabel("Resultado")
plt.ylabel("Frecuencia")
plt.show()
```

#### Binomial

`Ejemplo`: validar si nuestro __Value at Risk__ es consistente con nuestro nivel de seguridad.

```python
# Muestra (no de portafolios / activos que ocupan este tipo de VaR)
N = 100000
# DÃ­as observados resultados de VaR de un portafolio o activo.
n = 252
# Nivel de significancia
p = 0.05

muestra = stats.binom.rvs(n,p,size=N,random_state=None)

mu, sigma2 = stats.binom.stats(n, p, moments="mv")
print(f"Valores estimados")
print(f"E[X]:{mu}, V[X]:{sigma2}, sd[X]:{np.sqrt(sigma2)}")

plt.hist(muestra)
plt.title("DistribuciÃ³n Binomial")
plt.xlabel("Resultado")
plt.ylabel("Frecuencia")
plt.show()
```
